```python
f1 = figure(figsize=[fig_width/3.05, fig_width/4.1]) #halffig(hf=0.6)
ax1 = f1.add_axes([0,0,1,1])

f2 = figure(figsize=[fig_width/3.05, fig_width/4.1]) #halffig(hf=0.6)
ax2 = f2.add_axes([0,0,1,1])

f3 = figure(figsize=[fig_width/3.05, fig_width/4.1]) # halffig(hf=0.6)
ax3 = f3.add_axes([0,0,1,1])

f4 = figure(figsize=[fig_width/3.05, fig_width/4.1]) # halffig(hf=0.6)
ax4 = f4.add_axes([0,0,1,1])


w = ax1.bbox.bounds[2]
h = ax1.bbox.bounds[3]

axrange = np.array([-2.5, 5.5])
print 

for ax in [ax1, ax2, ax3, ax4]:
    ax.set_axis_off()
    ax.set_xlim(*axrange)
    ymax = axrange[0] + (axrange[1] - axrange[0]) * (h/w)
    ax.set_ylim(axrange[0], ymax)
    
    ax.arrow(0,0,axrange[1]-0.5,0,width=0.01,color='gray',length_includes_head=True, zorder=-1)
    ax.arrow(0,0,0,ymax-0.5,width=0.01,color='gray',length_includes_head=True, zorder=-1)
    
    ax.text(axrange[1]-0.6,0.2,r"$\parallel$", horizontalalignment='center', verticalalignment='bottom', color='gray')
    ax.text(+0.2,ymax-0.7,r"$\perp$", horizontalalignment='left', verticalalignment='center', color='gray')
    
    ax.plot(0,0,"ko", zorder=0)
    ax.add_patch(matplotlib.patches.Circle((0,0), 1, color="none", ec="black"))
    ax.text(0,-0.7,"$j$",verticalalignment='baseline', horizontalalignment='center')
    
    x = 2.1 if ax in [ax1, ax2] else 2.8
    ax.plot(x,0,"ko", zorder=0)
    ax.add_patch(matplotlib.patches.Circle((x,0), 1.4, color="none", ec="black"))
    ax.text(x,-0.7,"$i$",verticalalignment='baseline', horizontalalignment='center')
    
    dx = 2 if ax in [ax1, ax2] else -1.2
    dy = 1
    
    xar = linspace(-15, 15, 1000)
    ax.plot(xar, (xar-x) * (dy/dx), "k:", lw=0.5, dashes=(1,1), zorder=0)

    ax.arrow(x, 0, dx, dy, width=0.01, color='red', length_includes_head=True, zorder=1)
    if ax in [ax1, ax3]:
        ax.plot([x, x+dx, x+dx], [dy, dy, 0], color='red', linestyle=":", dashes=(1,1), lw=0.5)
        ax.arrow(x, 0, dx, 0, width=0.01, color='red', length_includes_head=True, zorder=1)
        ax.arrow(x, 0, 0, dy, width=0.01, color='red', length_includes_head=True, zorder=1)
        

        ax.text(x-0.2, 1.5, r"$u_{\perp,ij}$", color='red', verticalalignment='baseline', horizontalalignment='left')
        ax.text(x+dx-0.2, 1.5, r"$\vec{u_{ij}}$", color='red', verticalalignment='baseline', horizontalalignment='left')
        ax.text(x+dx-0.2, -0.7, r"$u_{\parallel,ij}$", color='red', verticalalignment='baseline', horizontalalignment='left')    
        
    if ax in [ax2, ax4]:
        # when do the circles touch?
        mindelta = sqrt(xar**2 + ((xar-x) * (dy/dx))**2) - 2.4
        ids = np.where(diff(sign(mindelta)) != 0)[0]

        #ax3 = ax.twinx()
        #ax3.plot(xar, )

        #ax3.set_xlim(*axrange)
        #ax3.set_ylim(-10,10)
        for i,linestyle in zip(ids, ["dotted", "dashed"]):
            xpos = xar[i]
            ypos = (xar[i] - x) * (dy/dx)
            ax.add_patch(matplotlib.patches.Circle((xpos,ypos), 1.4, color="none", ec="red", linestyle=linestyle, lw=2))
            ax.plot(xpos, ypos, "ro", zorder=0, mec="None")
   

ax1.text(0.02,0.98, r'(a)',
        transform=ax1.transAxes,
        verticalalignment='top',
        horizontalalignment='left',
        size='large')

ax2.text(0.02,0.98, r'(b)',
        transform=ax2.transAxes,
        verticalalignment='top',
        horizontalalignment='left',
        size='large')

ax3.text(0.02,0.98, r'(.)',
        transform=ax3.transAxes,
        verticalalignment='top',
        horizontalalignment='left',
        size='large')

ax4.text(0.02,0.98, r'(c)',
        transform=ax4.transAxes,
        verticalalignment='top',
        horizontalalignment='left',
        size='large')

f1.savefig("bk.pdf")
f2.savefig("bk2.pdf")
f3.savefig("mk.pdf")
f4.savefig("mk2.pdf")
```
