```python
t = np.arange(100)

s1 = np.zeros(t.shape)

msk = (t>20) * (t<60)
s1[msk] = (t[msk]-20)/40
s1[t>=60] = 1

s2 = np.zeros(t.shape)
s2[t>20] = 1

plt.plot(t,s1)
plt.plot(t,s2)

plt.axis(ymin=-0.1, ymax=1.1)
```

```python
fullfig()
plt.subplots_adjust(wspace=0.02, hspace=0.04)
ax1 = plt.subplot(2,2,1)
ax2 = plt.subplot(2,2,2)
ax3 = plt.subplot(2,2,3)
ax4 = plt.subplot(2,2,4)

ax1.axis(xmin=0, xmax=90, ymin=-0.1, ymax=1.1)
ax2.axis(xmin=0, xmax=90, ymin=-0.1, ymax=1.1)
ax3.axis(xmin=0, xmax=90, ymin=0, ymax=10)
ax4.axis(xmin=0, xmax=90, ymin=0, ymax=10)

ax1.set_xticks([0,20,40,60,80])
ax1.set_xticklabels([])
ax2.set_yticklabels([])
ax2.set_xticks([0,20,40,60,80])
ax2.set_xticklabels([])
ax3.set_xticks([0,20,40,60,80])
ax3.set_yticks([0,2,4,6,8])
ax4.set_yticklabels([])
ax4.set_xticks([20,40,60,80])
ax4.set_yticks([0,2,4,6,8])

ax1.set_ylabel("$s$")
ax3.set_ylabel(r"$\Delta t$")
ax3.set_xlabel("$t$")
ax4.set_xlabel("$t$")

ax1.plot(t, s1)
ax2.plot(t, s2)

s1_delta_array = np.array([(s1[dt:] - s1[:-dt])[:90] for dt in range(1,11)])
s2_delta_array = np.array([(s2[dt:] - s2[:-dt])[:90] for dt in range(1,11)])

ax3.imshow(s1_delta_array, vmin=0, vmax=1, extent=[0,90,1,10], interpolation='None', origin='lower', aspect='auto')
m = ax4.imshow(s2_delta_array, vmin=0, vmax=1, extent=[0,90,1,10], interpolation='None', origin='lower', aspect='auto')

cb = plt.colorbar(mappable=m, ax=[ax1,ax2,ax3,ax4])
cb.set_label(r"$\Delta s$", labelpad=0)
cb.set_ticks([0,1])

plt.savefig('delta-s-example.pdf')
```
