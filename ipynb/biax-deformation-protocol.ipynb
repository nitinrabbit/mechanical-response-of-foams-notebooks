```python
fullfig(hf=0.5125)

plt.subplots_adjust(left=0, right=1.05, top=0.95, bottom=0, wspace=0)
ax1 = plt.subplot(1,4,1)
ax2 = plt.subplot(1,4,2)
ax3 = plt.subplot(1,4,3)
ax4 = plt.subplot(1,4,4)

#print ax.bbox.height, ax.bbox.width

axes = [ax1, ax2, ax3, ax4]

for ax in axes:
    ax.set_ylim(-0.05,1.05)
    ax.set_xlim(-0.05,1.05)
    ax.axis('off')
    
Aj = 0.5
phij = 0.842
phi = 1

def box_coordinates(A, s):
    xmax = np.sqrt(A) * np.exp(s)
    ymax = np.sqrt(A) * np.exp(-s)
    return xmax, ymax

def xc(xmax):
    return [0,xmax,xmax,0,0]

def yc(ymax):
    return [0,0,ymax,ymax,0]

va = 'center' # crappy maar het beste

# plot 1: from some shape to s1=-0.2, A=Aj
ax = ax1
x1,y1 = box_coordinates(Aj, 0.1)
x2,y2 = box_coordinates(Aj, -0.2)
#ax.plot(xc(x1), yc(y1), color='gray', dashes=(1,1))
ax.plot(xc(x2), yc(y2), color='black')
ax.text(0,1,r'$A=A_j,~s_{CD}=-0.2$', fontsize='small', va=va)
#ax.annotate('', (x2,y2), (x1,y1),
#             arrowprops={
#        'arrowstyle': '->',
#        'linewidth': 0.5,
#        'shrinkA': 0,
#        'shrinkB': 3.5,
#        'connectionstyle': 'arc3, rad=-0.3'}
#)
subplotlabel('(a)', 'lower left', ax=ax)

# plot 2: from s1=-0.2, A=Aj to A=Aj*phifactor
ax=ax2
x1,y1 = box_coordinates(Aj, -0.2)
x2,y2 = box_coordinates(Aj*(phij/phi), -0.2)
ax.plot(xc(x1), yc(y1), color='gray', dashes=(1,1))
ax.plot(xc(x2), yc(y2), color='black')
ax.text(0,1,r'$A\rightarrow A_j \phi_j/\phi$', fontsize='small', va=va)
ax.annotate('', (x2*0.8,y2*0.8), (x1,y1),
             arrowprops={
        'arrowstyle': '->',
        'shrinkB':0,
        'shrinkA':0,
        'linewidth': 0.5,}
)
subplotlabel('(b)', 'lower left', ax=ax)

# plot 3: 
ax=ax3
x1,y1 = box_coordinates(Aj*(phij/phi), -0.2)
x2,y2 = box_coordinates(Aj*(phij/phi), 0.2)
ax.plot(xc(x1), yc(y1), color='gray', dashes=(1,1))
ax.plot(xc(x2), yc(y2), color='black')
t = ax.text(0,1,r'$s_{CD}\rightarrow +0.2$', fontsize='small', va=va)
ax.annotate('', (x2,y2), (x1,y1),
             arrowprops={
        'arrowstyle': '->',
        'linewidth': 0.5,
        'shrinkA': 0,
        'shrinkB': 5,
        'connectionstyle': 'arc3, rad=0.3'}
)
subplotlabel('(c)', 'lower left', ax=ax)

# plot 3: 
ax=ax4
x1,y1 = box_coordinates(Aj*(phij/phi), 0.2)
x2 = x1
y2 = Aj/x2

ax.plot(xc(x1), yc(y1), color='gray', dashes=(1,1))
ax.plot(xc(x2), yc(y2), color='black')
ax.text(0,1,r'$A\rightarrow 1.1 A_j$', fontsize='small', va=va)
ax.annotate('', (0.5*x2,y2*1.2), (0.5*x1,y1),
             arrowprops={
        'arrowstyle': '->',
        'shrinkA': 0,
        'shrinkB': 0,
        'linewidth': 0.5}
)
subplotlabel('(d)', 'lower left', ax=ax)

savefig('deformations.pdf')
```
