```python
# U vs delta

halffig()
ax1 = plt.subplot(211, aspect='equal')
ax2 = plt.subplot(212, aspect='equal')
ax1.axis([-1,1,0,1])
ax2.axis([-1,1,0,1])
plt.subplots_adjust(hspace=0)
ax1.set_xticklabels([])
ax1.set_xticks([0])
ax2.set_xticklabels([])
ax2.set_xticks([0])
ax1.set_ylabel(r"$\delta_{ij}$", rotation='horizontal')
ax2.set_ylabel(r"$U_{ij}$", rotation='horizontal')
ax1.spines['right'].set_visible(False)
ax1.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)
ax2.spines['top'].set_visible(False)

x = np.linspace(-1,1)
y1 = x.copy()*0.65
y1[x<0] = 0

y2 = y1**2 * 2

ax1.set_yticks([0])
ax2.set_yticks([0])

ax1.plot(x,y1,color='blue', lw=2, zorder=3, clip_on=False)
ax2.plot(x,y2,color='red', lw=2, zorder=3, clip_on=False)

for x in [-0.8, -0.4, 0, 0.4, 0.8]:
    circle=plt.Circle((x, 0.7), 0.1, color=(0,0,0,0.2),ec='k',clip_on=False,zorder=5)
    ax1.add_artist(circle)
    circle=plt.Circle((x, 0.9-0.1*x), 0.1,color=(0,0,0,0.2),ec='k',clip_on=False,zorder=5)
    ax1.add_artist(circle)
    

ax1.plot([0.8,0.95], [0.8,0.8], color='black', dashes=(1,1), lw=0.5)
ax1.plot([0.8,0.95], [0.9-0.1*0.8-0.1,0.9-0.1*0.8-0.1], color='black', dashes=(1,1), lw=0.5)
ax1.text(0.97, 0.75, r"$\delta_{ij}$", va='center', fontsize='x-small')

subplotlabel('(a)', ax=ax1, loc='lower left')
subplotlabel('(b)', ax=ax2, loc='lower left')

plt.savefig('delta-explanation.pdf')    
    
```
